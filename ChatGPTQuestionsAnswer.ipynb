```python
import docx
import spacy
import nltk

# Load the Docx file
doc = docx.Document('test.docx')

# Initialize an empty list to store the questions
questions = []

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Loop through all the paragraphs in the Docx file and extract potential questions
for paragraph in doc.paragraphs:
    # Parse the paragraph text with spaCy
    doc = nlp(paragraph.text)

    # Loop through all the sentences in the paragraph
    for sent in doc.sents:
        # Check if the sentence ends with a question mark
        if sent.text.strip().endswith('?'):
            questions.append(sent.text.strip())
        else:
            # Check if the sentence has a verb in the second position
            verb_in_second_position = (len(sent) > 2 and sent[2].pos_ == 'VERB')
            if verb_in_second_position:
                # Check if the sentence starts with a capital letter
                if sent.text.strip().capitalize() == sent.text.strip():
                    questions.append(sent.text.strip())

# Print the extracted questions
for question in questions:
    print(question)
```

```python
# Print the extracted questions
def is_question(question):
    #from nltk.tokenize import word_tokenize
# Tokenize the sentence into words
    words = nltk.word_tokenize(question)

    # Check if the first word is a question word or if the sentence ends with a question mark
    if words[0].lower() in ["what", "when", "where", "who", "whom", "whose", "which", "why", "how"] or question[-1] == "?":
        return True
    else:
        return False
```

```python
def get_bullet_points_in_table(doc, table_index):
    """
    Returns a list of all the bullet point paragraph text in a table in a Word document.

    Args:
        - doc (docx.Document): the Word document object
        - table_index (int): the index of the table in the document

    Returns:
        A list of all the bullet point paragraph text in the table.
    """
    table = doc.tables[table_index]
    bullet_points = []

    # Iterate over each cell in the table
    for row in table.rows:
        for cell in row.cells:
            # Check if the cell contains a paragraph with bullet points
            for p in cell.paragraphs:
                if p.style.name.startswith('List Bullet'):
                    # Get the text of each bullet point and append it to the list
                    for run in p.runs:
                        bullet_points.append(run.text)

    return bullet_points
```
